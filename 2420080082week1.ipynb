```python
import pandas as pd
import numpy as np
```

```python
student_data = pd.read_csv("student_data (2).csv")

print("Student Dataset Preview:")
display(student_data.head())

print("\nMissing values:")
print(student_data.isnull().sum())
```

```python
num_cols = student_data.select_dtypes(include=['int64','float64']).columns
cat_cols = student_data.select_dtypes(include=['object']).columns

# Fill numerical
for col in num_cols:
    student_data[col] = student_data[col].fillna(student_data[col].mean())

# Fill categorical
for col in cat_cols:
    student_data[col] = student_data[col].fillna(student_data[col].mode()[0])

student_data.drop_duplicates(inplace=True)

print("After cleaning:")
print(student_data.isnull().sum())
```

```python
titanic = pd.read_csv("Titanic-Dataset.csv")

print("Titanic Preview:")
display(titanic.head())

print("\nMissing BEFORE:")
print(titanic.isnull().sum())
```

```python
num_cols_t = titanic.select_dtypes(include=['int64','float64']).columns
cat_cols_t = titanic.select_dtypes(include=['object']).columns

for col in num_cols_t:
    titanic[col] = titanic[col].fillna(titanic[col].mean())

for col in cat_cols_t:
    titanic[col] = titanic[col].fillna(titanic[col].mode()[0])

print("\nMissing AFTER:")
print(titanic.isnull().sum())
```
